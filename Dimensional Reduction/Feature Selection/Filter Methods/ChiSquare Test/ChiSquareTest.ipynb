```python
import pandas as pd
X = pd.DataFrame({
    'gender': [0, 1, 0, 1, 1],
    'married': [1, 1, 0, 0, 1],
    'income': [20000, 40000, 50000, 30000, 60000]
})

y = [0, 1, 1, 0, 1]

X
```

```python
# Chi-Square requires non-negative values
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

selector = SelectKBest(score_func=chi2, k=2)

X_new = selector.fit_transform(X_scaled, y)

selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
import pandas as pd
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.preprocessing import MinMaxScaler

# sample dataset
X = pd.DataFrame({
    'gender': [0,1,0,1,1],
    'married': [1,0,1,0,1],
    'education': [1,0,1,1,0],
    'income': [5000,3000,4000,3500,6000],
    'property_area': [1,0,1,1,0],
    'favorite_color': [0,1,2,0,1]
})

y = [1,0,1,0,1]  # loan status

# scale (chi-square needs non-negative)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# chi-square selection
selector = SelectKBest(score_func=chi2, k=3)
X_new = selector.fit_transform(X_scaled, y)

# selected feature names
selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features.tolist())
```

1. Chi-Square test

2. Categorical (or non-negative numeric) features

3. Feature not related to target

4. Works only with categorical targets (classification)
