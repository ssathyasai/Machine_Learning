# **Variance Threshold**

```python
import pandas as pd
data=pd.DataFrame(
    {
        'age':[10,20,30,40,50],
        'salary':[1000,2000,3000,4000,5000],
        'country_code':[1,1,1,1,1],
    }
)
data
```

```python
from sklearn.feature_selection import VarianceThreshold
selector=VarianceThreshold(threshold=1)
selected_data=selector.fit_transform(data)
selected_features=data.columns[selector.get_support()]
print("Selected Features:",selected_features)
```

```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold

# create sample dataset
data = pd.DataFrame({
    "Age": [25, 45, 52, 23, 40],
    "Blood Pressure": [120, 140, 135, 118, 142],
    "Cholesterol Level": [200, 220, 210, 190, 215],
    "Gender": [1, 0, 1, 0, 1],
    "Country Code": ["India"] * 5,
    "Hospital ID": ["H001"] * 5,
    "Machine Version": ["v2.1", "v2.1", "v2.1", "v2.1", "v2.2"],
    "Disease": [0, 1, 1, 0, 1]
})

# encode categorical
data_enc = pd.get_dummies(data, drop_first=True)

# separate features
X = data_enc.drop(columns=["Disease"])

# apply variance filter
selector = VarianceThreshold(threshold=1)
X_selected = selector.fit_transform(X)

# get selected feature names
selected_features = X.columns[selector.get_support()]

print("Selected Features:", selected_features)
```
