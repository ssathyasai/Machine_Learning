```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# create dataset
data = pd.DataFrame({
    'size_sqft': [1000,1500,2000,2500,3000],
    'tiles_count': [100,150,200,250,300],
    'price': [200000,300000,400000,500000,600000]
})

data

# compute correlation matrix
correlation_matrix = data.corr()
print(correlation_matrix)

# visualize correlation
sns.heatmap(correlation_matrix, annot=True)
plt.show()

# high correlation  multicollinearity problem

# removing highly correlated features (>0.9)
upper = correlation_matrix.where(
    np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
)

to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
print("Highly Correlated Features to drop:", to_drop)

# drop redundant features (e.g., tiles_count)
data_reduced = data.drop(columns=to_drop)
print("Remaining columns:", data_reduced.columns.tolist())
```

```python
# High correlation causes multicollinearity (duplicate information)

# Remove Number of Tiles and Construction Cost (redundant with House Size)

# Keeping them leads to unstable coefficients and overfitting risk

# Use correlation-based filter (Pearson correlation)
```
